```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = '/content/images elephant.jpg'
```

```python
original_image_bgr = cv2.imread(image_path)
if original_image_bgr is None:
  print(f"Failed to read image file: {image_path}")
else:

  original_image_rgb = cv2.cvtColor(original_image_bgr, cv2.COLOR_BGR2RGB)
```

```python
negative_image_bgr = cv2.bitwise_not(original_image_bgr)
negative_image_rgb = cv2.cvtColor(negative_image_bgr, cv2.COLOR_BGR2RGB)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(original_image_rgb)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(negative_image_rgb)
axes[1].set_title('Negative Image (OpenCV)')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
contrast_image_bgr = cv2.convertScaleAbs(original_image_bgr, alpha=1.5, beta=0)
contrast_image_rgb = cv2.cvtColor(contrast_image_bgr, cv2.COLOR_BGR2RGB)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(original_image_rgb)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(contrast_image_rgb)
axes[1].set_title('Contrast Image')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

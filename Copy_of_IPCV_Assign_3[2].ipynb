```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/lion gray.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
if img is None:
    print("Error: Image not found.")
else:

    hist = np.zeros(256, dtype=int)

    print("Image loaded and histogram array initialized.")
```

```python
height, width = img.shape
for i in range(height):
    for j in range(width):
        pixel_value = img[i, j]
        hist[pixel_value] += 1
```

```python
total_pixels = img.size

pdf = hist / total_pixels
cdf = np.zeros(256, dtype=float)
cumulative_sum = 0

for i in range(256):
    cumulative_sum += pdf[i]
    cdf[i] = cumulative_sum
```

```python
new_value = np.zeros(256, dtype=int)
for i in range(256):
    new_value[i] = round(255 * cdf[i])

equalized_img = np.zeros_like(img)
for i in range(height):
    for j in range(width):
        p = img[i, j]
        equalized_img[i, j] = new_value[p]
```

```python
def process_histogram_equalization(image_path):

    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if img is None:
        print(f"Error: Image not found at {image_path}.")
        return


    hist = np.zeros(256, dtype=int)
    height, width = img.shape
    for i in range(height):
        for j in range(width):
            pixel_value = img[i, j]
            hist[pixel_value] += 1


    total_pixels = img.size
    pdf = hist / total_pixels
    cdf = np.zeros(256, dtype=float)
    cumulative_sum = 0
    for i in range(256):
        cumulative_sum += pdf[i]
        cdf[i] = cumulative_sum


    new_value = np.zeros(256, dtype=int)
    for i in range(256):
        new_value[i] = round(255 * cdf[i])


    equalized_img = np.zeros_like(img)
    for i in range(height):
        for j in range(width):
            p = img[i, j]
            equalized_img[i, j] = new_value[p]


    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(equalized_img, cmap='gray')
    plt.title('Histogram Equalized Image')
    plt.axis('off')

    plt.show()

print("Function 'process_histogram_equalization' defined successfully.")
```
